# Processed match schema

This notebook defines and validates the schema contract for cleaned MatchCast data.

```python
PROCESSED_MATCH_COLUMNS = (
    "match_id",
    "source_row_number",
    "date",
    "home_team",
    "away_team",
    "original_home_team",
    "original_away_team",
    "home_score",
    "away_score",
    "tournament",
    "city",
    "country",
    "neutral",
    "result",
    "goal_difference",
)

PROCESSED_MATCH_DTYPES = {
    "match_id": "string",
    "source_row_number": "Int64",
    "date": "datetime64[ns]",
    "home_team": "string",
    "away_team": "string",
    "original_home_team": "string",
    "original_away_team": "string",
    "home_score": "Int64",
    "away_score": "Int64",
    "tournament": "string",
    "city": "string",
    "country": "string",
    "neutral": "boolean",
    "result": "string",
    "goal_difference": "Int64",
}

RESULT_VALUES = frozenset({"H", "D", "A"})
```

## Validation

These checks keep the schema internally consistent and verify nullable types, traceability fields, and result values.

```python
assert len(PROCESSED_MATCH_COLUMNS) == len(set(PROCESSED_MATCH_COLUMNS))
assert set(PROCESSED_MATCH_COLUMNS) == set(PROCESSED_MATCH_DTYPES)

assert PROCESSED_MATCH_DTYPES["home_score"] == "Int64"
assert PROCESSED_MATCH_DTYPES["away_score"] == "Int64"
assert PROCESSED_MATCH_DTYPES["neutral"] == "boolean"

traceability_fields = {
    "match_id",
    "source_row_number",
    "original_home_team",
    "original_away_team",
}
assert traceability_fields <= set(PROCESSED_MATCH_COLUMNS)
assert RESULT_VALUES == {"H", "D", "A"}

print("Processed match schema validation passed.")
```
